## Transform features:

```python
train = pd.read_csv("Datasets/train_dataset_hackathon_mkb.csv", sep = ';', encoding = 'cp1251')
test = pd.read_csv("Datasets/test_dataset_hackathon_mkb.csv", sep = ';', encoding = 'cp1251')
```

```python
uniqs = train['WORKERSRANGE'].unique()
replacements = dict()

for i, j in zip([un for un in uniqs if type(un) == str], [int(un.split()[0]) for un in uniqs if type(un) == str]):
    
    replacements[i] = j
```

```python
float_cols = train.columns[train.dtypes == 'float64']
int_cols = train.columns[train.dtypes == 'int64']

F_cols = [col for col in float_cols if col[:2] == "F1" or col[:2] == "F2"]
non_F_cols = [col for col in float_cols if col[:2] != "F1" and col[:2] != "F2"]
```

```python
def transformations(df):
    
    for col in F_cols:
        
        df[col] = np.log10(df[col])
        
    drop_cols = ['id_contract',
                'SIGN_DATE', 
                'DATEFIRSTREG',
                'DATEFIRSTREG',
                'TAXREGPAY_REGDATE',
                'OKVED_CODE',
                'BIRTHDATE',
                'OKTMO_CODE',
                'TAXREG_REGDATE'
                ]
    
    df['WORKERSRANGE'].replace(replacements, inplace = True)
    
    df.drop(columns = drop_cols, inplace = True)
    
    df['SEX_NAME'].replace({"" : 0, "" : 1}, inplace = True)
    
    ohe_cols = [
        'OKFS_GROUP',
        'OKOPF_GROUP',
        'OKOGU_GROUP'
    ]
    
    for col in ohe_cols:
        
        df = df.join(pd.get_dummies(df[col]))
        df[col] = df[col].isna().astype(int)
        
    col_name = ' '    
    df[col_name] = (df['CITIZENSHIP_NAME'] == col_name).astype(int)
    
    df.drop(columns = 'CITIZENSHIP_NAME', inplace = True)
    
    return df
```

```python
train = transformations(train)
test = transformations(test)
```

```python
target = np.array(train['TARGET'])
train.drop(columns = 'TARGET', inplace = True)
```

## Catboost

```python
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold
```

```python
from catboost import CatBoostClassifier
```

```python
params = {
    'n_estimators' : 500,
    'max_depth' : 12,
}
```

```python
models = []
k = 1

for train_ids, test_ids in StratifiedKFold(n_splits=8).split(train, target):
    
    X_train, X_test = train.loc[train_ids], train.loc[test_ids]
    Y_train, Y_test = target[train_ids], target[test_ids]
    
    model = CatBoostClassifier(**params)
    model.fit(X_train, Y_train, eval_set = (X_test, Y_test), early_stopping_rounds = 25)
    
    print(f"Iter number {k}", metrics.roc_auc_score(Y_test, model.predict_proba(X_test)[:, 1]))
    models.append(model)
    k += 1
```

```python
test2 = pd.read_csv("Datasets/test_dataset_hackathon_mkb.csv", sep = ';', encoding = 'cp1251')
```

```python
test2['TARGET'] = 0.0

for model in models:
    
    test2['TARGET'] += model.predict_proba(test)[:, 1]
    
test2['TARGET'] /= 8
```

```python
test2[['id_contract', 'TARGET']].to_csv("submit_kfold.csv", sep = ';', index = False)
```
